```python
data = pd.read_csv("train_values_wJZrCmI.csv")
labels = pd.read_csv('train_labels.csv')
data = data.merge(labels, on='row_id')
print(data) #import data
```

```python
data.drop('bank_interest_rate',axis=1,inplace=True)
data.drop('mm_interest_rate',axis=1,inplace=True)
data.drop('mfi_interest_rate',axis=1,inplace=True)
data.drop('other_fsp_interest_rate',axis=1,inplace=True) #dropping columns with majority nulls values
```

```python
data['education_level'].fillna(data['education_level'].mean())
data['share_hh_income_provided'].fillna(data['share_hh_income_provided'].mean()) #filling null values with mean
```

```python
from sklearn.preprocessing import LabelEncoder
le_rel = LabelEncoder()
data['religion'] = le_rel.fit_transform(data['religion'])

le_c = LabelEncoder()
data['country'] = le_c.fit_transform(data['country'])

le_rel_to_hh = LabelEncoder()
data['relationship_to_hh_head'] = le_rel_to_hh.fit_transform(data['relationship_to_hh_head'])

le_ecly = LabelEncoder()
data['employment_category_last_year'] = le_ecly.fit_transform(data['employment_category_last_year'])
 
le_etly = LabelEncoder()
data['employment_type_last_year'] = le_etly.fit_transform(data['employment_type_last_year'])  #label encoding
```

```python
data['country'] = data['country'].astype('str')
```

```python
mode0 = data[data['country'] == '0']['education_level'].mode()
mode1 = data[data['country'] == '1']['education_level'].mode()
mode2 = data[data['country'] == '2']['education_level'].mode()
mode3 = data[data['country'] == '3']['education_level'].mode()
mode4 = data[data['country'] == '4']['education_level'].mode()
mode5 = data[data['country'] == '5']['education_level'].mode()
mode6 = data[data['country'] == '6']['education_level'].mode() 
```

```python
def fill_ed(data):
    education = data[0]
    country = data[1]

    if pd.isnull(education):
        if country is '0': 
            return 2
        elif country is '1': 
                return 0
        elif country is '2': 
                return 1
        elif country is '3': 
                return 2
        elif country is '4': 
                return 2
        elif country is '5': 
                return 2
        else: 
                return 1
    else:
        return education
```

```python
data['education_level'] = data[['education_level','country']].apply(fill_ed, axis=1)
```

```python
data['education_level'].unique()
```

```python
data['share_hh_income_provided'] = data['share_hh_income_provided'].fillna(0)
```

```python
categorical = ['phone_ownership', 'employment_type_last_year', 'employment_category_last_year', 'education_level', 'relationship_to_hh_head', 'religion', 'country', 'share_hh_income_provided']
data[categorical] = data[categorical].astype('category')
```

```python
cat_dummies = pd.get_dummies(data[categorical], drop_first = True)
data_dum = data.drop(categorical, axis = 1)
data = pd.concat([data_dum, cat_dummies], axis =1)
```

```python
data_1 = data.drop(data.columns[-1],axis=1)
labels_1 = labels.drop(['row_id'],axis=1)
```

```python
data_1.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data_1,labels_1,test_size=0.20) #spliting into 80/20
```

```python
threshold = 0.5
y_train_binary = (y_train > threshold).astype(int)
y_test_binary = (y_test>threshold).astype(int)  #converting probablity to 0 or 1
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train_binary) #logistic regression
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
accuracy = precision_score(y_test_binary, y_predict)
print(accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier() #decisionn tree
dt.fit(X_train,y_train_binary )
```

```python
y_pred_dt = dt.predict(X_test)
accuracy_dt = precision_score(y_test_binary,y_pred_dt)
print(accuracy_dt)
```

```python
conf_matrix = confusion_matrix(y_test_binary,y_pred_dt)
print(conf_matrix)
```

```python
from sklearn.ensemble import RandomForestClassifier  #random forest classifier
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train_binary)
```

```python
y_pred_rf = rf_classifier.predict(X_test)
accuracy_rf = precision_score(y_test_binary,y_pred_dt)
print(accuracy_rf)
```
